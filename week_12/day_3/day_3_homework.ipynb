```python
# Question 2

# Join the data containing card details and customer details by customer id, so that all records of card details and any matching records in customer details are kept.

card.merge(customer, how='left', on='Cust_ID')

```

```python
# Question 3

# Join the data containing fraud details with transaction details so all rows of both tables are kept. What does the resulting row number tell you?

fraud.merge(transaction, how = 'outer', on = 'Transaction_ID')

```

```python
# Question 4 - [Harder] Join the data containing card details with transaction details so rows from the first which have matching ones in the second are returned.

pd.merge(card, transaction, left_on='Card_Number', right_on='Credit_Card_ID')


```
